**Q9. Chart Selection Justification**<br>
**Explain why different charts were used for different business questions in this analysis.**

- **Line Chart (Sales Trend & Growth):** 
  - Used to show sales over time because trends, seasonality, and growth patterns are best understood through continuous time-based visuals.
- **Bar Chart (Category Performance):** 
  - Ideal for comparing performance across categories since it clearly shows differences in total sales.
- **Heatmap (Category vs Region):** 
  - Helps quickly identify high and low-performing regions using color intensity, making comparisons faster.
- **Pie Chart (Regional Contribution):** 
  - Shows percentage contribution of each region to total sales.
- **Scatter Plot (Profitability Investigation & Product Prioritization):** 
  - Used to analyze the relationship between sales and profit and to detect high-profit/high-sales products.

**Q10. Business Insight Summary**<br>
**Summarize two key insights from your dashboard and explain how they can help management make better decisions.**
- **Technology category generates the highest sales:**
  - Management should prioritize inventory, marketing, and expansion strategies in this category to maximize revenue.

- **Certain regions contribute significantly more to overall sales:**
  - Management can focus investments in high-performing regions while developing strategies to improve weaker regions.
