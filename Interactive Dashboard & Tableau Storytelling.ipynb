## Interactive Dashboard & Tableau Storytelling

**Q1 Explain the difference between a Tableau Dashboard and a Tableau Story Point. Give two healthcare-related use cases.**
- **Tableau Dashboard:** A single screen that displays multiple related visualizations together. It is interactive and used for real-time monitoring and analysis.
- **Tableau Story Point:** A sequence of visualizations arranged in order to present a narrative or step-by-step analysis.

- **Healthcare Use Cases:**
  - **Dashboard:** Monitor ER wait times, ICU bed availability, and staff on duty for daily hospital operations.
  - **Story Point:** Present a disease outbreak analysis showing affected areas, age groups, and recommended vaccination zones.

<br>

**Q2 What are Actions in Tableau? Explain how Filter Action can help hospital management analyze patient data.**
- **Actions:** Interactive features that trigger a response (like filtering or highlighting) when users click or hover on data.

- **Filter Action:** When a user selects data in one chart, other charts update to show only related data.

- **Hospital Example:**
  - If management clicks on Cardiology, all other charts update to show only Cardiology data helping them analyze performance and identify issues quickly.
