### Publishing and Sharing in Tableau

**Q1. What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/Cloud in terms of accessibility and data security?**
- The key difference is accessibility and security.

  - In Tableau Public, dashboards are publicly accessible to anyone on the internet, and the data cannot be kept private.

  - In Tableau Server or Tableau Cloud, dashboards are restricted to authorized users, and administrators can control access using permissions and authentication.


<br>

**Q2. When publishing a Tableau dashboard, what happens to a live connection compared to an extract? How does this affect data refresh and performance?**
- A live connection keeps the dashboard directly connected to the original data source. This means the dashboard always displays the most up-to-date data, but performance may depend on the speed and availability of the underlying database.

- An extract stores a snapshot of the data within Tableau. The data does not update automatically and requires manual or scheduled refreshes. However, extracts often improve performance because the data is optimized and stored in Tableau’s internal format.


<br>

**Q3: In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users can view, interact with, or edit a published dashboard?**
- The feature used to control user access is Permissions in Tableau Server or Tableau Cloud.

- Permissions allow administrators to define what users or groups can do with published content such as dashboards, workbooks, or data sources. Through permissions, admins can control actions like viewing, interacting, editing, downloading, or publishing.


<br>

**Q4: If you want users to view a dashboard but not download the underlying data, which permission setting would you adjust in Tableau Server?**
- You would adjust the Download Data permission.

- To enforce this:
  - Allow the View permission so users can see and interact with the dashboard.
  - Deny or disable the Download Data permission to prevent them from downloading the underlying dataset.


<br>

**Q5: Name two methods by which a Tableau dashboard can be embedded into a web page or website.**
- Two common methods to embed a dashboard created in Tableau into a website are:

- Embed Code (HTML iframe):
  - After publishing a dashboard to Tableau Server, Tableau Cloud, or Tableau Public, you can use the generated HTML embed code (iframe) and place it inside a webpage’s HTML.

- APIs:
  - Developers can use the Tableau JavaScript Embedding API to integrate dashboards into websites with more control, such as custom interactions, filtering, and events.


<br>

**Q6: List two best practices you should follow when preparing a dashboard for public sharing to ensure data security and privacy.**
- Remove sensitive or confidential data:
  - Ensure that personal, confidential, or proprietary information is removed before publishing the dashboard publicly.

- Review permissions and data access:
  - Carefully configure user permissions and access controls when publishing on Tableau Server or Tableau Cloud so that only authorized users can view or interact with the dashboard.


<br>

**Q7: After publishing a dashboard, users report a “Data not found” error. What is the most common cause related to data source settings?**
- The most common cause is a broken or inaccessible data source connection after publishing.

- When a dashboard is published to Tableau Server or Tableau Cloud, the original data source path or credentials may not be available on the server.
  - The server cannot access the local file path used in Tableau Desktop.
  - Database credentials were not embedded or configured correctly.
  - The data source was not published along with the workbook.


<br>

**Q8: When embedding a Tableau dashboard into a secure internal portal, what two additional security measures should be implemented to ensure only authorized users can access the embedded content?**
- Two important security measures are:

  - User Authentication:
    - Ensure users must authenticate through Tableau Server or Tableau Cloud (such as SSO, Active Directory, or login credentials) before accessing the embedded dashboard.

  - Proper Permission Settings:
    - Configure permissions and user groups so only authorized users can view or interact with the dashboard.
