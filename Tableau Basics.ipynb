**Q4. What insight can be gained by comparing Sales and Profit together? Then explain how you would create a chart to analyze Profit by Sub-Category.**

- Comparing these two metrics helps you identify Profitability Gaps.
  - **High Sales, Low Profit:** This suggests high overhead costs, aggressive discounting, or high shipping expenses. It tells you that while the product is popular, it isn't making the company money.
  - **Low Sales, High Profit:** This indicates "niche" high-margin items that are very efficient but might need more marketing to scale.
  - **Negative Profit (Loss):** This highlights "bleeding" areas where every sale actually costs the company money, which is a red flag for immediate action.

- Creating a "Profit by Sub-Category" Chart:
  - **Select the Dimension:** From the Data Pane on the left, locate Sub-Category (under the Orders table). Drag and drop it onto the Columns shelf at the top.
  - **Select the Measure:** Locate Profit (the green # field) in the Data Pane. Drag and drop it onto the Rows shelf.
  - **Automatic Visualization:** Tableau will automatically generate a Bar Chart showing the total Profit for each Sub-Category.
  - **Refine or Sort:** Click the "Sort Descending" icon in the toolbar to see the most profitable items at the top.

**Q7. Explain the business importance of geographical analysis. Then describe how you would create a Map showing Sales by State using Sample – Superstore.**
- Geographical analysis is crucial because it reveals spatial patterns that traditional tables hide:
  - **Performance Hotspots:** It identifies which specific regions or states are driving the most revenue.
  - **Strategic Allocation:** Companies use it to decide where to focus marketing budgets or open new physical locations.
  - **Logistics Optimization:** It helps in understanding customer density to improve shipping routes and supply chain efficiency.

- How to Create a Sales Map:
  - **Initialize Geography:** Double-click the State field in the Data Pane; Tableau will automatically place Longitude in Columns and Latitude in Rows.
  - **Add Value:** Drag the Sales measure from the Data Pane and drop it onto the Size or Color marks card.

**Q8. What is the difference between a Worksheet and a Dashboard in Tableau? Why is a dashboard more useful for decision-makers?**
- Difference Between Worksheet & Dashboard:
  - **Worksheet:** A single workspace used to create one specific chart or graph (e.g., "Sales by date" line chart).
  - **Dashboard:** A combined view that displays multiple worksheets on a single screen.

- Why Dashboards are better for Decision-Makers:
  - **The Big Picture:** They allow leaders to see sales, profit, and maps all at once instead of clicking through different tabs.
  - **Interactivity:** Clicking a data point in one chart (like a specific State) can filter all other charts on the dashboard instantly.
  - **Speed to Insight:** They highlight correlation like a dip in profit happening only in a specific region—allowing for faster business decisions.

**Q9.  What is Tableau Prep, and why is it important in real-world data projects?** <br>
**Then explain how you would:** <br>
**Remove null values** <br>
**Remove duplicate records** <br>
**Change incorrect data types (Answer in steps, no tool access required.)** 
- Tableau Prep is a tool used to combine, shape, and clean "messy" data before analysis. It is vital in real-world projects because data is often inconsistent, and Prep ensures your final dashboard is accurate and reliable.

- Cleaning Steps:
  - **Remove Null Values:** Click the field containing nulls, right-click the "null" value in the profile pane, and select Exclude.
  - **Remove Duplicate Records:** Add an Aggregate step, group by all relevant fields, and Tableau will collapse identical rows into one.
  - **Change Data Types:** Click the Data Type icon (e.g., # or Abc) above the column header and select the correct type (e.g., Date or String) from the list.
