## Calculated Fields, Parameters & LOD Expressions

### Theoretical Questions 

**Q1: What is a Calculated Field and why is it important in business dashboards?**
- A calculated field is a custom formula created from existing data in Tableau.
- It’s important because it creates KPIs (like profit margin), applies business logic, and turns raw data into meaningful insights.

<br>

**Q2: Why can’t parameters change visuals directly in Tableau?**
- Parameters are just input values. They don’t affect visuals unless used inside calculated fields, filters, or reference lines. They need logic to drive changes.

<br>

**Q3: When should EXCLUDE LOD be used?**
- Use EXCLUDE when you want to remove a dimension from the view’s level of detail—commonly for percent-of-total or comparing detail vs summary values.

<br>

**Q4: Difference between FIXED and INCLUDE LOD expressions?**
- FIXED: Locks calculation at specified dimensions, ignores the view.
- INCLUDE: Adds a dimension to the current view’s level of detail.

- Simple rule:
  - FIXED = lock level
  - INCLUDE = add level
